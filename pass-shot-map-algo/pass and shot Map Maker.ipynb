#SHOT MAP


```python
# Inputing title and csv file directory

title = input('Enter Shot Map Title: ')
CSVdir = input('Enter File Directory: ')

# Transformaing csv to dataframe

data = pd.read_csv(CSVdir)


# Drawing the pitch

pitch = Pitch(pitch_type='opta',
              pitch_length=105, pitch_width=68, pitch_color='#0f1029', line_color='White')
fig, ax = pitch.draw()

# Plotting the data depending on the outcome

for i in range(len(data['Event'])):
    if data['Event'][i] == 'Goal':
        plt.scatter(data["X"][i],data["Y"][i], s=100,  c="Green")
    elif data['Event'][i] == 'Shot on Goal':
        plt.scatter(data["X"][i],data["Y"][i], s=100,  c="#2bffea")
    else:
        plt.scatter(data["X"][i],data["Y"][i], s=100,  c="Red")


# Below is all styling and adding text


fig.suptitle(title, fontsize=20, color="White", fontweight='bold', fontfamily="Monospace")
fig.set_size_inches(15, 7)
fig.set_facecolor("#0f1029")

fig.text(.06,.1,'Shot on Goal', size=18, color='#2bffea', fontweight='bold', fontfamily="Monospace")
fig.text(.06,.3,'Goal', size=18, color='Green', fontweight='bold', fontfamily="Monospace")
fig.text(.06,.2,'Missed Shot', size=18, color='Red', fontweight='bold', fontfamily="Monospace")
ax.set_xlabel('Twitter: @sakusaif_zer0', fontfamily="Monospace", color = "White")


plt.show()
```

#PASS MAP

```python
# Inputing title and csv file directory


title2 = input('Enter Pass Map Title: ')
CSVdir2 = input('Enter CSV file Directoy: ')

# Transformaing csv to dataframe

data2 = pd.read_csv(CSVdir2)



# Drawing the pitch

pitch = Pitch(pitch_type='opta', orientation = 'horizontal',
              pitch_length=105, pitch_width=68, pitch_color='#0f1029', line_color='White')

fig, ax = pitch.draw()


# Plotting the data deoending on the outcome


for i in range(len(data2['Event'])):
    if data2['Event'][i] == 'Pass':
        plt.plot([data2["X"][i], data2["X2"][i]], [data2["Y"][i], data2["Y2"][i]], color="Green", linewidth= 2.0)
        plt.scatter(data2["X"][i],data2["Y"][i], s=100,  color="Green", marker='x')
    elif data2['Event'][i] == 'Assist':
        plt.plot([data2["X"][i], data2["X2"][i]], [data2["Y"][i], data2["Y2"][i]], color="#2bffea", linewidth= 2.0)
        plt.scatter(data2["X"][i],data2["Y"][i], s=100,  color="#2bffea", marker='x')

    else:
        plt.plot([data2["X"][i], data2["X2"][i]], [data2["Y"][i], data2["Y2"][i]], color="Red", linewidth= 2.0)
        plt.scatter(data2["X"][i],data2["Y"][i], s=100,  color="Red", marker='x')


# Below is all styling and adding text


fig.suptitle(title2, fontsize=14, color="White", fontweight='bold', fontfamily="Monospace")
fig.set_facecolor("#0f1029")

fig.text(.06,.1,'Assist', size=18, color='#2bffea', fontweight='bold', fontfamily="Monospace")
fig.text(.06,.3,'Successful pass', size=14, color='Green', fontweight='bold', fontfamily="Monospace")
fig.text(.06,.2,'Missed pass', size=18, color='Red', fontweight='bold', fontfamily="Monospace")
ax.set_xlabel('Twitter: @sakusaif_zer0', fontfamily="Monospace", color = "White")
fig.set_size_inches(15, 7)
plt.gca().invert_yaxis()


plt.show()



```
