```python
# Importing Modules
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.pyplot import figure
import scipy.misc
from scipy import ndimage
```

```python
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Lucida Sans'
```

```python
# Loading images

stroke_wht = mpimg.imread('white-brush-stroke.png')
stroke_blk = mpimg.imread('black-brush-stroke.png')
logo = mpimg.imread('w_mark.png')
bg_element = mpimg.imread('bg_element.png')
```

```python
def template_expat_report(player_name, bg_img_dir, stat_1, num_1, stat_2, num_2, stat_3, num_3, stat_4, num_4):
    # Loading the bg img
    player_img = mpimg.imread(bg_img_dir)

    # Set image size in inches
    figure(figsize=(15, 12))
    ax = plt.gca()

    # Set axis limit
    ax.set_xlim(0, 1500)
    ax.set_ylim(0, 1400)

    # Show the images, use extent to change size and coordinates
    imgplot = plt.imshow(bg_element, extent=(0, 1500, 0, 1400))
    imgplot = plt.imshow(logo, extent=(1250, 1500, 0, 250), alpha = 0.5)
    imgplot = plt.imshow(player_img, extent=(-500, 1800, -100, 1400), alpha = 0.5)

    # Brush Strokes
    imgplot = plt.imshow(stroke_wht, extent=(1150, 1500, 900, 1100))
    imgplot = plt.imshow(stroke_wht, extent=(1150, 1500, 700, 900))
    imgplot = plt.imshow(stroke_wht, extent=(1150, 1500, 500, 700))
    imgplot = plt.imshow(stroke_wht, extent=(1150, 1500, 300, 500))
    imgplot = plt.imshow(stroke_blk, extent=(350, 1200, 1200, 1450))

    # Labels made invisible
    ax.set_xlabel('', color='Black', fontsize = 13, fontfamily="sans-serif", fontweight = 'bold', alpha = 0)
    ax.set_ylabel('', color='Black', fontsize = 13, fontfamily="sans-serif", alpha = 0)
    
    #####################################################################################################################
    
    plt.text(100, 950, stat_1, color='White', fontsize=30, fontfamily="sans-serif", fontweight = 'bold')
    plt.text(1200, 950, num_1, color='red', fontsize=32, fontfamily="sans-serif", fontweight = 'bold')
    
    plt.text(100, 750, stat_2, color='White', fontsize=30, fontfamily="sans-serif", fontweight = 'bold')
    plt.text(1200, 750, num_2, color='red', fontsize=32, fontfamily="sans-serif", fontweight = 'bold')
    
    plt.text(100, 550, stat_3, color='White', fontsize=30, fontfamily="sans-serif", fontweight = 'bold')
    plt.text(1200, 550, num_3, color='red', fontsize=32, fontfamily="sans-serif", fontweight = 'bold')

    plt.text(100, 350, stat_4, color='White', fontsize=30, fontfamily="sans-serif", fontweight = 'bold')
    plt.text(1200, 350, num_4, color='red', fontsize=32, fontfamily="sans-serif", fontweight = 'bold')

    plt.text(500, 1300, player_name, color='White', fontsize=40, fontfamily="sans-serif", fontweight = 'bold')
    

    # General image style (Background...etc.)
    ax.set_facecolor("darkred")
    ax.spines['right'].set_color('darkred')
    ax.spines['top'].set_color('darkred')
    ax.spines['left'].set_color('darkred')
    ax.spines['bottom'].set_color('darkred')
    plt.tick_params(axis='x', labelsize=0, color='White')
    plt.tick_params(axis='y', labelsize=0, color='White')
    plt.show()
    
    
template_expat_report('Wahbi Khazri', 'khazri.jpeg', "Tirs Cadrés", '21/30', 'Passes clés', '33', 'Penalty réussis', '4/4', 'Dribbles réussis', '33/67')
```
