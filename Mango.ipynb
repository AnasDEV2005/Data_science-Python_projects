```python
import matplotlib.pyplot as plt
from highlight_text import fig_text
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import re
```

```python
d = pd.read_csv('Mango2.csv')
d.fillna(0)

d
```

```python
filtered_df = d[d['Price'].notnull()]
filtered_df

```

```python


fig, ax = plt.subplots(figsize =(10, 7))

plt.bar(filtered_df['Type'], filtered_df['Pack'], width=0.1, color='Orange')
plt.text(0, 35, 'Total Number of Boxes: 577', fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)
plt.xlabel('Type', fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)
plt.ylabel('Number of Boxes', fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_color('Red')
ax.spines['bottom'].set_color('Red')
plt.tick_params(axis='x', labelsize=10, color='White')
plt.tick_params(axis='y', labelsize=13, color='White')
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(10, 7))


plt.plot(filtered_df['Date'], filtered_df['Pack'], color = 'Orange')
plt.text(0, 35, 'In Chronological Order', fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)
plt.xlabel('From 14 July to 17 August', fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)
plt.ylabel('Number of Boxes', fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_color('Red')
ax.spines['bottom'].set_color('Red')
plt.tick_params(axis='x', labelsize=0, color='White')
plt.tick_params(axis='y', labelsize=13, color='White')
plt.show()
```

```python
total_gain = filtered_df['Price'].sum()


fig, ax = plt.subplots(figsize =(14, 10))
string='Total Gain = '+ str(int(total_gain)) + ' ' + '.Â£'
plt.bar(filtered_df['Type'],  filtered_df['Price'], color = 'Orange')
plt.text(0, 7000, string, fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)
plt.xlabel('Type', fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)
plt.ylabel("Gain (EGY Pounds)", fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_color('Red')
ax.spines['bottom'].set_color('Red')
plt.tick_params(axis='x', labelsize=13, color='White')
plt.tick_params(axis='y', labelsize=13, color='White')
plt.show()
```

```python
from scipy.ndimage.filters import gaussian_filter1d



    
Average_gain = total_gain/577


fig, ax = plt.subplots(figsize =(15, 7))
string='Average Gain = '+ str(int(Average_gain)) + ' ' + 'egy Â£'




daf = filtered_df.copy(); daf['gain_per_box'] = daf['Price']/daf['Pack']



ysmoothed = gaussian_filter1d(daf['gain_per_box'], sigma=5)


plt.plot(  filtered_df['Date'], ysmoothed, color = 'Orange')




plt.text(0, 230, string, fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)
plt.xlabel('Type', fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)
plt.ylabel("Gain per Box", fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.5)


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_color('Red')
ax.spines['bottom'].set_color('Red')
plt.xticks(range(0, len(filtered_df['Date']), 3))
plt.tick_params(axis='x', labelsize=13, color='White')
plt.tick_params(axis='y', labelsize=13, color='White')
plt.show()
```
