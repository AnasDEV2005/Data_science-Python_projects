Importing the Modules

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

Importing the data used to train Model.

```python
df = pd.read_csv('model_training_data.csv')
df.head(30)
```

Data description

* Bodypart: 
 - Headed Shot = 0
 - Foot Shot = 1
* Outcome:
 - Goal = 1
 - Not Goal = 0
* X & Y:
 - From 0 to 100

Importing Logistic Regression Module

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

Training the model

```python
gen1 = model.fit(df[['X','Y','Bodypart']], df[['Outcome']])


data2 = pd.read_csv('train_angles.csv')

data2

gen1.fit(df[['X','Y','Bodypart']], df[['Outcome']])
```

Here the model will calculate goal probability (xG)

```python
d = pd.read_csv('events.csv')
d
xg = gen1.predict_proba(d[['X','Y','Bodypart']])
xgoal = []
for i in range(len(d['X'])):
    xgoal.append(xg[i][1])
    


total_goals = len(d['X'])
    
    
d['goal_xg'] = xgoal
total_xg = d['goal_xg'].sum()

d
```

Finally: The Visualization (Here Test)

```python
pitch = Pitch(pitch_type='opta', pitch_color='#2e2c4d', line_color='White', half = True, orientation='Vertical')
fig, axs = pitch.draw(figsize=(10, 8))
for i in range(len(d['X'])):
    color = "#ff4433"
    if d['Outcome'][i] == 1:
        color = "#33ff69"
        pitch.scatter(d['X'][i], d['Y'][i], s=((d['goal_xg'][i])*500),marker = 'o',  c=color, ax=axs, alpha=0.5)
    else:
        pitch.scatter(d['X'][i], d['Y'][i], s=((d['goal_xg'][i])*500),  c=color, ax=axs, alpha=0.5)

plt.gca().invert_yaxis()

goals = 'Total Goals = '+ str(total_goals)
wm = "@StatsKoora"
axs.set_xlabel(goals, color='White', fontsize = 20, fontfamily="sans-serif")
axs.set_ylabel(wm, color='White', fontsize = 20, fontfamily="sans-serif", alpha = 0.35, fontweight='bold')




fig.suptitle('TEST GOAL MAP ', fontsize=20, color="White", fontweight='bold', fontfamily="sans-serif")
fig.set_size_inches(15, 7)
plt.text(60, 15, 'Size = Goal Probability', fontsize=18, fontfamily='sans-serif', fontweight = 'bold', alpha = 0.7, color='White')
fig.set_facecolor("#2e2c4d")
```
